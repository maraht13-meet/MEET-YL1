The dataset I am analyzing is about hot dog types. It came from http://lib.stat.cmu.edu/DASL/Datafiles/Hotdogs.html and the information about the dataset came from http://lib.stat.cmu.edu/DASL/Datafiles/Hotdogs.html. I want to investigate a relationship between hot dog calories and hot dog sodium.


```python
import csv
```

```python
f = open('food.csv', 'rb')
reader = csv.reader(f) 
calories = []
sodium = []
for row in reader :
    a=row[0]
    b = a.split()
    c = b[1::]
    d = map(int , c)
    calories.append(d[0])
    sodium.append(d[1])
    
    
print calories
print sodium 
    
    
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()                                                                                                
m = fig.add_subplot(111)
m.set_title('Calories and Sodium in Hot Dogs')
m.set_xlabel('Calories')
m.set_ylabel('Sodium')
m.scatter(calories, sodium)
fig.savefig("calories_and_sodium_with_titles.png")
```

This scatter plot shows a complicated relationship between hot dog sodium and and hot dog calories.
